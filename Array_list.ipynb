que1 : find Duplicates 

```python
arr = [1,2,3,2,4,3,5,6]
seen = set()
duplicate = set()

for i in arr:
    if i in seen:
        duplicate.add(i)
    else:
        seen.add(i)
duplicate = list(duplicate)
print("duplicate elements are:",duplicate)
```

que2: Merge list

```python
lst1 = [1,4,6]
lst2 = [3,2,7]

i = j = 0
merge = []
while i < len(lst1) and j < len(lst2):
    if lst1[i] < lst2[j]:
        merge.append(lst1[i])
        i += 1
    else:
        merge.append(lst2[j])
        j += 1
        
merge +=lst1[i:]
merge +=lst2[j:]
print(merge)
```

que3 : Frequency Count


```python
lst = [1, 2,1, 3,2,5,6,1,7]
freq = {}
for i in lst:
    if i not in freq:
        freq[i] = 1
    else:
        freq[i] += 1
        
print(freq)
```

que4:rotate array by k steps 

```python
arr = [1,2,3,4,5]
k = 3
# using slicing 
new_arr = arr[-k:] + arr[:-k]
print(new_arr)
```

que5: find missing number


```python
arr = [1,2,4,5,6]
total = sum(arr)
n = len(arr)+1
exp_sum = n * (n+1) // 2

m_n = exp_sum - total
print(m_n)

```

que6 : Largest element in list

```python
lst = [1,2,3,8,5,78,2,99]
largest = lst[0]
for i in range(len(lst)):
    if lst[i] > largest:
        largest = lst[i]
print(largest)    
```
