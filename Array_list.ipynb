que1 : find Duplicates 

```python
arr = [1,2,3,2,4,3,5,6]
seen = set()
duplicate = set()

for i in arr:
    if i in seen:
        duplicate.add(i)
    else:
        seen.add(i)
duplicate = list(duplicate)
print("duplicate elements are:",duplicate)
```

que2: Merge list

```python
lst1 = [1,4,6]
lst2 = [3,2,7]

i = j = 0
merge = []
while i < len(lst1) and j < len(lst2):
    if lst1[i] < lst2[j]:
        merge.append(lst1[i])
        i += 1
    else:
        merge.append(lst2[j])
        j += 1
        
merge +=lst1[i:]
merge +=lst2[j:]
print(merge)
```
