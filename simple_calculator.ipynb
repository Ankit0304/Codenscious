Simple Calculator using Arithmetic Operator 
(/, * , -, +)

```python
number1 = int(input("enter a first number:"))
print("Your first number:",number1)
number2 = int(input("enter a second number:"))
print("Your Second number:",number2)
addition = number1 + number2
subtraction = number1 - number2 
mul = number1 * number2
div = number1 / number2
fdiv = number1 // number2 
exp = number1 ** number2

print(addition, subtraction, mul, div, fdiv, exp)
```
