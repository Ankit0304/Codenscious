Basic Operations

```python
fruits = {'apple', 'banana', 'mango', 'grapes','lichi'}
print('set is:', fruits)
fruits.add('gavava')
print("after adding", fruits)
fruits.remove('lichi')
print('after removing:',fruits)
fruits.discard('strawberry')
print('new set is:',fruits)


```

Union Set

```python
set1 = {10,20,30,40,50}
set2 = {30,60,70,40}

# new_set = set1 | set2
new_set = set2.union(set1)
print(new_set)
```
