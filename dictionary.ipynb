Basic Operations

```python
my_dict = {'name': 'Ankit', 'age':23, 'language':'Python'}
print('Original Dict', my_dict)
my_dict['passout'] = 2025
print('updated dict', my_dict)
my_dict['age'] = 21
print('modify age', my_dict)

del my_dict['language']
print('after deleting language', my_dict)

print("key value pairs")
for key,value in my_dict.items():
    print(f'{key}:{value}')

# empty whole dict
# my_dict.clear() 

new_dict = {"person":{"name":"ankit", "age":21}}
res = new_dict["person"]['age']
print("nested age", res)

#check values
n_dict = {"a":2, "b":3, "c":4, "d":5}
if 3 in n_dict.values():
    print("3 in dict")
else:
    print("3 is not in list")

#update key
new_dict["User"] = new_dict.pop("person")
print("change key", new_dict)
```
