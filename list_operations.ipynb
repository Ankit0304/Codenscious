que 1: basic operations

```python
lst = [1,3, 5, 6, 3, 2,'Ankit', True]
print("length of list:", len(lst))
print("third element:", lst[2])
if len(lst) == 0:
    print("List is empty")
else:
    print("List is not empty")
    
print("original list:", lst)
lst[1] = 200
print("after updating 2nd element:", lst)

lst.append(400)
print("after appending 400:", lst)

lst.insert(3, 600)
print("after inserting 600 at index 3:", lst)

lst.pop()
print("after popping last element:", lst)

lst.remove(5)
print("after removing element 5:", lst)

# lst.sort()
# print("after sorting the list:", lst)

lst.reverse()
print("after reversing the list:", lst)

```
