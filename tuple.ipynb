que 1 : tuple operations

```python
tup = (1, 2, 3, 4, 5, 6, 7)
print("length of tuple:", len(tup))
print("third element of tuple:", tup[2])

# tuple repetition
rep_tup = tup*2
print("repeated tuple:", rep_tup)

#  slicing
sliced_tup = tup[3:]
print("sliced tuple from index 3:", sliced_tup)

# reverse
rev = tup[::-1]
print("reversed tuple:", rev)

# nested tuple 
tup2 = (2, 3, (4,5,6),[2,4],'orange')
res = tup2[2][2]
print("element from nested tuple:", res)

# type conversion
lst = [2,4,6]
print(type(lst))
tup3 = tuple(lst)
print("converted tuple:", tup3)
print(type(tup3))
lst = list(tup3)
print("converted list:", lst)
print(type(lst))

```
