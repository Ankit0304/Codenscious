que 1

# que 2


```python
# pattern printing
n = int(input("Enter a number:"))
for i in range(1, n+1):
    for j in range(1,i+1):
        print(j, end=" ")
    print('')
```

 que 3: sum of all number upto given number

```python
n = int (input("Enter a number:"))
sum = 0
for i in range(n+1):
    sum += i
print(sum)
```

que 4 Pattern reverse

```python
n = int(input("enter a number:"))
for i in range(n,0,-1):
    for j in range(i,0,-1):
        print(j, end='')
    print('')
```

que 5: pattern


```python
n = int(input("enter a number:"))
for i in range(1,n+1):
    for j in range(1,i+1):
        print('*', end='')
    print('')
for i in range(n-1,0,-1):
    for j in range(i,0,-1):
        print('*', end='')
    print('')
```
