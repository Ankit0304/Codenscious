que1 :Count Vowels

```python
String = "heytherehowwhenwhere"
vowels = 'aeiou'
String = String.lower()
count = 0
for i in String:
    if i in vowels:
        count += 1
    
print(count)
```

que2 : check Palindrome


```python
def is_palindrome(s):
    reverse_word = s[::-1]
    if s == reverse_word:
        print("string is palindrome")
    else:
        print("string is not palindrome")

s = input("enter the string:")
is_palindrome(s)    
```
