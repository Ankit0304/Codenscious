que1 :Count Vowels

```python
String = "heytherehowwhenwhere"
vowels = 'aeiou'
String = String.lower()
count = 0
for i in String:
    if i in vowels:
        count += 1
    
print(count)
```

que2 : check Palindrome


```python
def is_palindrome(s):
    reverse_word = s[::-1]
    if s == reverse_word:
        print("string is palindrome")
    else:
        print("string is not palindrome")

s = input("enter the string:")
is_palindrome(s)    
```

que3 : check Anagram

```python
def is_anagram(s1,s2):
    if len(s1) != len(s2):
        return False
    
    s1 = sorted(s1.lower())
    s2 = sorted(s2.lower())
    
    if s1 == s2 :
        print("string are anagram")
    else:
        print("strings are not anagram")
    
    
s1 = 'greek'
s2 = 'kreeG'
is_anagram(s1,s2)
```

que4 : word frequency count

```python
def count_freq(text):
    words = text.split()
    freq = {}
    for i in words:
        if i not in freq:
            freq[i] = 1
        else:
            freq[i] += 1
    return freq

text = input("enter the text:")
count_freq(text)   
```
