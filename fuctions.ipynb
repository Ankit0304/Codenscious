que1 : Factorial 


```python
def fact(n):
    if n==0 or n==1:
        return 1
    else:
        for i in range (2,n):
            n = n*i
        return n
    
num = int(input("Enter a number:"))
print(fact(num))
```

que2 : fibonacci series

```python
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=' ')
        a, b = b, a+b

num = int(input("Enter the number of terms:"))
fibonacci(num)
```

que3: lambda function


```python
square = lambda x: x*x 
num = int(input("Enter a number:"))
print(square(num))
```

que 4: Generate even num in a given range

```python

def generate_even(start, end):
    lst = []
    for i in range(start, end+1):
        if i%2 == 0:
            lst.append(i)
    return lst

start = int(input("Enter start of range:"))
end = int(input("Enter end of range:"))
print(generate_even(start, end))
```

que 5: sum of digit using recursion

```python

def sum_of_digits(n):
    if n == 0:
        return 0 
    else: 
        return n % 10 + sum_of_digits(n // 10)
    
n = int(input('enter a number:'))  
print(sum_of_digits(n))
```
