que1 : Factorial 


```python
def fact(n):
    if n==0 or n==1:
        return 1
    else:
        for i in range (2,n):
            n = n*i
        return n
    
num = int(input("Enter a number:"))
print(fact(num))
```

que2 : fibonacci series

```python
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=' ')
        a, b = b, a+b

num = int(input("Enter the number of terms:"))
fibonacci(num)
```
